```python
import os
import shutil
```

```python
def split_pneumonia(folder):
    pneumonia_path = f"../dataset/chest_xray/{folder}/PNEUMONIA"
    bacteria_path = f"../dataset/chest_xray/{folder}/BACTERIA"
    virus_path = f"../dataset/chest_xray/{folder}/VIRUS"

    os.makedirs(bacteria_path, exist_ok=True)
    os.makedirs(virus_path, exist_ok=True)

    for file in os.listdir(pneumonia_path):
        src = os.path.join(pneumonia_path, file)

        if "bacteria" in file.lower():
            dst = os.path.join(bacteria_path, file)
        elif "virus" in file.lower():
            dst = os.path.join(virus_path, file)
        else:
            continue

        shutil.move(src, dst)

    print(f"{folder} done!")
```

```python
split_pneumonia("train")
split_pneumonia("test")
split_pneumonia("val")
```
